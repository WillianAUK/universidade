#### a) F√≥rmula de Machin

```python
def piMachin(x):
    return (4*np.ùëéùëüùëêùë°ùëéùëõ( 1/5))/x - (np.ùëéùëüùëêùë°ùëéùëõ( 1/239)/x)
```

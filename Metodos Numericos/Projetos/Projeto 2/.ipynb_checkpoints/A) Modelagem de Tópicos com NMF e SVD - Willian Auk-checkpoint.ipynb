Considere o caso mais extremo - reconstruir a matriz usando um produto externo de dois vetores. Claramente, na maioria dos casos, não seremos capazes de reconstruir a matriz com exatidão. Mas se tivéssemos um vetor com a frequência relativa de cada palavra do vocabulário na contagem total de palavras e um com o número médio de palavras por documento, então esse produto externo seria o mais próximo possível.

Agora considere aumentar essas matrizes para duas colunas e duas linhas. A decomposição ideal agora seria agrupar os documentos em dois grupos, cada um dos quais tendo uma distribuição de palavras tão diferente quanto possível entre si, mas o mais semelhante possível entre os documentos no agrupamento. Chamaremos esses dois grupos de 'tópicos'. E agruparíamos as palavras em dois grupos, com base nas que aparecem com mais frequência em cada um dos tópicos.

Grandes quantidades de dados são coletados todos os dias. À medida que mais informações se tornam disponíveis, torna-se difícil acessar o que procuramos. Portanto, precisamos de ferramentas e técnicas para organizar, pesquisar e compreender grandes quantidades de informações.

A modelagem de tópicos nos fornece métodos para organizar, compreender e resumir grandes coleções de informações textuais. Ajuda em:

- Descobrir padrões tópicos ocultos que estão presentes em toda a coleção
- Anotae documentos de acordo com esses tópicos
- Usar essas anotações para organizar, pesquisar e resumir textos

A modelagem de tópicos pode ser descrita como um método para localizar um grupo de palavras (ou seja, tópico) de uma coleção de documentos que melhor representa as informações na coleção. Também pode ser pensado como uma forma de mineração de texto - uma maneira de obter padrões recorrentes de palavras em material textual.

Pegaremos um conjunto de dados de documentos em várias categorias diferentes e encontraremos tópicos (consistindo em grupos de palavras) para eles. Saber as categorias reais nos ajuda a avaliar se os tópicos que encontramos fazem sentido.

Tentaremos fazer isso com duas fatorações de matriz diferentes: Decomposição de Valor Singular (SVD) e Fatoração de Matriz Não-negativa (NMF)

### SVD - singular value decomposition
Em álgebra linear, a decomposição em valores singulares ou singular value decomposition (SVD) é a fatoração de uma matriz real ou complexa, com diversas aplicações importantes em processamento de sinais e estatística.

Formalmente, a decomposição em valores singulares de uma matriz m×n real ou complexa M é uma fatoração ou fatorização na forma:

$M = UΣV*$,

onde U é uma matriz unitária m×m real ou complexa, Σ é uma matriz retangular diagonal m×n com números reais não-negativos na diagonal, e V* (a conjugada transposta de V) é uma matriz unitária n×n real ou complexa. As entradas diagonais Σi,i de Σ são os chamados valores singulares de M. As m colunas de U e as n colunas de V são os chamados vetores singulares à esquerda e vetores singulares à direita de M, respetivamente.

A decomposição em valores singulares e a decomposição em autovalores são intimamente relacionadas. Mais especificamente:

- Os vetores singulares à esquerda de M são autovetores de $MM*$.
- Os vetores singulares à direita de M são autovetores de $M*M$.
- Os valores singulares não-nulos de M (ao longo da diagonal de Σ) são as raízes quadradas dos autovalores não-nulos de $M*M$ ou $MM*$.

Dentre as aplicações que envolvem a SVD estão o cálculo da pseudoinversa, o ajuste (fitting) de dados por mínimos quadrados, aproximação de matrizes, e a determinação do posto, imagem e núcleo de uma matriz.

### NMF - Non-negative matrix factorization

Fatoração de matriz não negativa (NMF ou NNMF), também aproximação de matriz não negativa  é um grupo de algoritmos em análise multivariada e álgebra linear onde uma matriz V é fatorada em (geralmente) duas matrizes W e H, Com a propriedade de que todas as três matrizes não têm elementos negativos. Essa não negatividade torna as matrizes resultantes mais fáceis de inspecionar. Além disso, em aplicações como processamento de espectrogramas de áudio ou atividade muscular, a não negatividade é inerente aos dados considerados. Uma vez que o problema não é exatamente resolvível em geral, ele é comumente aproximado numericamente.

O NMF encontra aplicações em campos como astronomia, visão computacional, agrupamento de documentos, imputação de dados perdidos, quimiometria, processamento de sinal de áudio, sistemas de recomendação e bioinformática.

- Função para plotar os graficos das palavras finais por NMF.

```python
def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7) 
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
```

- Função que processa os tópicos relevantes.

```python
num_top_words=12
def show_topics(a):
    top_words = lambda t: [vocab[i] for i in np.argsort(t)[:-num_top_words-1:-1]]
    topic_words = ([top_words(t) for t in a])
    return [' '.join(t) for t in topic_words]
```

- Funções complementares 

```python
mu = 1e-6
def grads(M, W, H):
    R = W@H-M
    return R@H.T + penalty(W, mu)*lam, W.T@R + penalty(H, mu)*lam 
```

```python
def penalty(M, mu):
    return np.where(M>=mu,0, np.min(M - mu, 0))
```

```python
def upd(M, W, H, lr):
    dW,dH = grads(M,W,H)
    W -= lr*dW; H -= lr*dH
```

```python
def report(M,W,H): 
    print(np.linalg.norm(M-W@H), W.min(), H.min(), (W<0).sum(), (H<0).sum())
```

- Funções para SVD

```python
def randomized_range_finder(A, size, n_iter=5):
    Q = np.random.normal(size=(A.shape[1], size))
    
    for i in range(n_iter):
        Q, _ = linalg.lu(A @ Q, permute_l=True)
        Q, _ = linalg.lu(A.T @ Q, permute_l=True)
        
    Q, _ = linalg.qr(A @ Q, mode='economic')
    return Q
```

```python
def randomized_svd(M, n_components, n_oversamples=10, n_iter=4):
    n_random = n_components + n_oversamples
    Q = randomized_range_finder(M, n_random, n_iter)
    B = Q.T @ M
    Uhat, s, V = linalg.svd(B, full_matrices=False)
    del B
    U = Q @ Uhat
    return U[:, :n_components], s[:n_components], V[:n_components, :]
```

#### A seguir, uma pequena demonstração de comparação com gráficos para demonstrar os resultados com NMF

- delimitando parâmetros

```python
n_samples = 3000
n_features = 1500
n_components = 10
n_top_words = 20
```

- Carregando dados

```python
print("Carregando dados...")
t0 = time()
data, _ = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes'), return_X_y=True)
data_samples = data[:n_samples]
print("demorou %0.3fs." % (time() - t0))
```

- Extraindo recursos tf-idf para NMF

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=n_features, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(data_samples)
print("demorou  %0.3fs." % (time() - t0))
```

- Extraindo recursos tf para LDA

```python
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=n_features, stop_words='english')
t0 = time()
tf = tf_vectorizer.fit_transform(data_samples)
print("demorou %0.3fs." % (time() - t0))
```

- Ajustando o modelo NMF (norma Frobenius) com recursos tf-idf, n_samples = 3000 e n_features = 1500

```python
t01 = time()
nmf = NMF(n_components=n_components, random_state=1, alpha=.1, l1_ratio=.5).fit(tfidf)
t1 = (time() - t01)
```

- Ajustando o modelo NMF (divergência Kullback-Leibler generalizada) com recursos tf-idf, n_samples = 3000 e n_features = 1500

```python
t02 = time()
nmf = NMF(n_components=n_components, random_state=1,
          beta_loss='kullback-leibler', solver='mu', max_iter=1000, alpha=.1,
          l1_ratio=.5).fit(tfidf)
t2 = (time() - t02)
```

- Adaptando modelos LDA com recursos tf, n_samples = 3000 e n_features = 1500

```python
lda = LatentDirichletAllocation(n_components=n_components, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
t03 = time()
lda.fit(tf)
t3 = (time() - t03)
```

- tempo de execução pra cada modelo:

```python
print("demorou %0.3fs." % t1, "o modelo NMF (norma Frobenius);")
print("demorou %0.3fs." % t2, "o modelo NMF (divergência Kullback-Leibler generalizada);")
print("demorou %0.3fs." % t3, "o modelo LDA; ")
```

- Plotando os gráficos para uma comparação

```python
tfidf_feature_names = get_feature_names_out()
plot_top_words(nmf, tfidf_feature_names, n_top_words, 'Tópicos no modelo NMF (norma Frobenius)')
```

```python
tfidf_feature_names = get_feature_names_out()
plot_top_words(nmf, tfidf_feature_names, n_top_words, 'Tópicos no modelo NMF (divergência Kullback-Leibler generalizada)')
```

```python
tf_feature_names = get_feature_names_out()
plot_top_words(lda, tf_feature_names, n_top_words, 'Tópicos no modelo LDA')
```
